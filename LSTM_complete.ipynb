```python
import pandas as pd
df = pd.read_csv('dataset.csv')
df.head()
```

```python
df.set_index('date')[['index','avg price', 'sales vol', 'cpi', 'crime count', 'VIX', 'bba mortgage rate']].plot(subplots=True, figsize=(10,10))
```

```python
df_input = df[['index', 'avg price', 'sales vol', 'cpi', 'crime count', 'VIX', 'bba mortgage rate']]
df_input
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
data_scaled = scaler.fit_transform(df_input)
data_scaled
```

```python
features = data_scaled
target = data_scaled[:,0]
```

```python
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
TimeseriesGenerator(features, target, length=12, sampling_rate=1, batch_size=1)[0]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.33, shuffle=False)
X_train.shape
```

```python
X_test.shape
```

```python
win_length = 1
batch_size = 1
num_features = 7
```

```python
train_generator = TimeseriesGenerator(X_train, y_train, length=win_length, sampling_rate=1, batch_size=batch_size)
test_generator = TimeseriesGenerator(X_test, y_test, length=win_length, sampling_rate=1, batch_size=batch_size)
train_generator[0]
```

```python
import tensorflow as tf
model = tf.keras.Sequential()
model.add(tf.keras.layers.LSTM(128, input_shape=(win_length, num_features), return_sequences=True))
model.add(tf.keras.layers.LeakyReLU(alpha =0.5))
model.add(tf.keras.layers.LSTM(128, return_sequences=True))
model.add(tf.keras.layers.LeakyReLU(alpha = 0.5))
model.add(tf.keras.layers.Dropout(0.3))
model.add(tf.keras.layers.LSTM(64, return_sequences=False))
model.add(tf.keras.layers.Dropout(0.3))
model.add(tf.keras.layers.Dense(1))
```

```python
model.summary()
```

```python
early_stopping = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=30, restore_best_weights=True, mode ='min')
model.compile(loss = tf.losses.MeanSquaredError(),
              optimizer = tf.optimizers.Adam(),
              metrics = [tf.metrics.MeanAbsoluteError()])
history = model.fit(train_generator, epochs=100, validation_data=test_generator, shuffle = False, callbacks=[early_stopping])
```

```python
model.evaluate(test_generator, verbose = 0)
```

```python
predictions = model.predict(test_generator)
predictions.shape[0]
```

```python
df_pred=pd.concat([pd.DataFrame(predictions), pd.DataFrame(X_test[:,1:][win_length:])], axis=1)
rev_trans = scaler.inverse_transform(df_pred)
rev_trans
```

```python
df_final = df_input[predictions.shape[0]*-1:]
```

```python
df_final['Pred']= rev_trans[:,0]
df_final
df_final[['index', 'Pred']].plot()
```

```python
# Create a TimeseriesGenerator for the entire dataset
full_generator = TimeseriesGenerator(features, target, length=win_length, sampling_rate=1, batch_size=batch_size)

# Generate predictions for the entire dataset
full_predictions = model.predict(full_generator)

# Revert the scaling for predictions and original values
df_full_pred = pd.concat([pd.DataFrame(full_predictions), pd.DataFrame(features[win_length:, 1:])], axis=1)
rev_trans_full = scaler.inverse_transform(df_full_pred)

# Add predictions to the original dataset
df_final_full = df_input[win_length:]
df_final_full['Pred'] = rev_trans_full[:, 0]

# Plot the actual vs. predicted values for the entire time series
df_final_full[['index', 'Pred']].plot(figsize=(12, 6), title='Actual vs. Predicted HPI for Entire Time Series')
```
